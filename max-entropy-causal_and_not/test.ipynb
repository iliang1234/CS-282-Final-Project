```python
import pandas as pd
```

```python
grads = pd.read_csv('all_grads.csv')
rewards = pd.read_csv('all_rewards.csv')
```

```python
filename = 'all_grads.txt'

# Open the file for writing
with open(filename, 'w') as file:
    file.write("[")
    for row in grads:
        file.write("[")
        for i, value in enumerate(row):
            if i < len(row) - 1:
                file.write(f"{value}, ")
            else:
                file.write(f"{value}")
        file.write("],\n")
    file.write("]")
```
